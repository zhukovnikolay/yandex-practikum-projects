Подготовим функции для исследования признаков

```python
# функция построения гистограммы с медианой, средним и модой
def my_hist(data, column, bins=10):
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.hist(data[column], bins=bins, alpha=0.8)
    ax.set_title(f'Гистограмма признака {column}')
    if data[column].dtype in ['float', 'int']: 
        feat_median = data[column].median()
        feat_mean = data[column].mean()
        feat_mode = data[column].mode()[0]
        ax.axvline(feat_median, color='g', linestyle='--', label='медиана')
        ax.axvline(feat_mean, color='r', linestyle='--', label='среднее')
        ax.axvline(feat_mode, color='y', linestyle='--', label='мода')
        ax.legend()
    plt.show()
```

```python
# функция, позволяющая исследовать признак
def feature_stat(data, column, need_hist=False, hist_bins=10, print_nan=False):
    missed_count = data[column].isna().sum()
    feat_mode = data[column].mode()[0]
    feat_stats = data[column].describe()
    if data[column].dtype not in ['object', 'bool']:
        print(f'Признак {column}\n{100 * "_"}\n'
              f'Кол-во пропусков (доля): {missed_count} ({missed_count / data.shape[0]:.2%})\n'
              f'Кол-во уникальных значений: {data[column].nunique()}\n'
              f'Минимальное значение: {feat_stats["min"]}\n'
              f'Максимальное значение: {feat_stats["max"]}\n'
              f'Медиана: {feat_stats["50%"]}\n'
              f'Среднее: {feat_stats["mean"]}\n'
              f'Мода: {feat_mode}\n'
              f'Тип данных: {data[column].dtype}\n' + 100 * '_')
    else:
        print(f'Признак {column}\n{100 * "_"}\n'
              f'Кол-во пропусков (доля): {missed_count} ({missed_count / data.shape[0]:.2%})\n'
              f'Кол-во уникальных значений: {data[column].nunique()}\n'
              f'Мода: {feat_mode}\n'
              f'Частотность моды: {feat_stats["freq"]}\n'
              f'Тип данных: {data[column].dtype}\n' + 100 * '_')
    if need_hist:
        try:
            print('Гистограмма')
            my_hist(data, column, bins=hist_bins)
            print(100 * '_')
        except:
            print('Не удалось построить гистограмму')
    if print_nan:
        print(f'Случайные объекты с пропусками')
        try:
            display(data[data[column].isna()].sample(min(5, data[column].isna().sum())))
        except:
            print('При выполнении произошла ошибка. Возможно, в данных нет пропусков')
        print(100 * '_')
```

Нам необходимо рассчитать суммарные продажи игры во всех регионах, также для дальнейшего исследования дополнительно создадим две таблицы с долей продаж для каждой платформы и каждого жанра в определенном году.

```python
# формируем признак summary_sales
data['summary_sales'] = data[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
```

```python
# проверяем
data.head(3)
```

Признак создан.

Создадим дополнительные информационные таблицы с долей продаж для каждой игровой платформы и каждого игрового жанра в каждом году.

```python
# определяем долю продаж для каждой игровой платформы в каждом году.
platforms_market_share = data.groupby(['platform', 'year_of_release'])['summary_sales'].sum().reset_index()
platforms_market_share = platforms_market_share.merge(platforms_market_share.groupby('year_of_release')['summary_sales'].sum().reset_index(), on='year_of_release')
platforms_market_share['market_share'] = platforms_market_share['summary_sales_x'] / platforms_market_share['summary_sales_y']
platforms_market_share = platforms_market_share.rename(columns={'summary_sales_x': 'summary_sales'})
platforms_market_share = platforms_market_share.drop(columns=['summary_sales_y'])
platforms_market_share
```

```python
# определяем долю продаж для каждого жанра в каждом году.
genres_market_share = data.groupby(['genre', 'year_of_release'])['summary_sales'].sum().reset_index()
genres_market_share = genres_market_share.merge(genres_market_share.groupby('year_of_release')['summary_sales'].sum().reset_index(), on='year_of_release')
genres_market_share['market_share'] = genres_market_share['summary_sales_x'] / genres_market_share['summary_sales_y']
genres_market_share = genres_market_share.rename(columns={'summary_sales_x': 'summary_sales'})
genres_market_share = genres_market_share.drop(columns=['summary_sales_y'])
genres_market_share
```

Низкую корреляцию оценок критиков с продажами для менее популярных платформ (PSV) можно объяснить их слабым аналитическим покрытием в связи с меньшей популярностью.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и альтернативную гипотезы:

$H_0:$ Средние пользовательские рейтинги платформ Xbox One и PC одинаковые ($rating_{Xbox} = rating_{PC}$).

$H_1:$ Средние пользовательские рейтинги платформ Xbox One и PC различаются ($rating_{Xbox} \neq rating_{PC}$).

```python
# определим размеры выборок
xbox_user_scores = filtered_data.query('platform == "XOne" & ~user_score.isna()')
pc_user_scores = filtered_data.query('platform == "PC" & ~user_score.isna()')
print(f'Размер выборки для Xbox One: {xbox_user_scores.shape[0]}\nДля PC: {pc_user_scores.shape[0]}')
```

Используем ttest_ind, т.к. сравниваем средние двух независимых выборок, аргумент equal_var установим в False, т.к. наши выборки разного размера (оценок для PC на 30% меньше).

Проведем тест при уровне значимости 0.01.

```python
alpha = 0.01

xbox_pc_user_scores_test_result = scipy.stats.ttest_ind(xbox_user_scores['user_score'], 
                                                        pc_user_scores['user_score'],
                                                        equal_var=False)

print('p-значение: ', xbox_pc_user_scores_test_result.pvalue)

if xbox_pc_user_scores_test_result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 

print(f'\nВыборочные средние:\n'
      f'Для Xbox One: {xbox_user_scores["user_score"].mean()}\n'
      f'Для PC: {pc_user_scores["user_score"].mean()}')
```

В данном случае не получилось отвергнуть нулевую гипотезу: средние оценки пользователей для платформ Xbox One и PC одинаковы (выборочные средние 6.59 и 6.29 соответственно).

Построим совмещенную гистограмму для наглядности.

```python
sns.distplot(xbox_user_scores['user_score'], label='Пользовательские оценки для Xbox One', bins=xbox_user_scores['user_score'].nunique()).set_title('Распределение пользовательских оценок для платформ Xbox One и PC');
sns.distplot(pc_user_scores['user_score'], label='Пользовательские оценки для PC', bins=xbox_user_scores['user_score'].nunique());
plt.axvline(xbox_user_scores['user_score'].mean(), color='b', linestyle='--', label='Средняя оценка для Xbox One');
plt.axvline(pc_user_scores['user_score'].mean(), color='r', linestyle='--', label='Средняя оценка для PC');
plt.legend();
```

На гистограмме видно, что распределения в целом схожи.

### Средние пользовательские рейтинги жанров Action и Sports разные.

Сформулируем нулевую и альтернативную гипотезы:

$H_0:$ Средние пользовательские рейтинги жанров Action и Sports одинаковые ($rating_{Action} = rating_{Sports}$).

$H_1:$ Средние пользовательские рейтинги жанров Action и Sports различаются ($rating_{Action} \neq rating_{Sports}$).

```python
# определим размеры выборок
action_user_scores = filtered_data.query('genre == "Action" & ~user_score.isna()')
sports_user_scores = filtered_data.query('genre == "Sports" & ~user_score.isna()')
print(f'Размер выборки для Action: {action_user_scores.shape[0]}\nДля Sports: {sports_user_scores.shape[0]}')
```

Используем ttest_ind, т.к. сравниваем средние двух независимых выборок, аргумент equal_var установим в False, т.к. наши выборки разного размера (оценок для Action более чем вдвое больше).

Проведем тест при уровне значимости 0.01.

```python
alpha = 0.01

genres_user_scores_test_result = scipy.stats.ttest_ind(action_user_scores['user_score'], 
                                                       sports_user_scores['user_score'],
                                                       equal_var=False)

print('p-значение: ', genres_user_scores_test_result.pvalue)

if genres_user_scores_test_result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 

print(f'\nВыборочные средние:\n'
      f'Для Action: {action_user_scores["user_score"].mean()}\n'
      f'Для Sports: {sports_user_scores["user_score"].mean()}')
```

В данном случае отвергам нулевую гипотезу и подтверждаем альтернативную: средние оценки пользователей для игр жанров Action и Sports различаются (выборочные средние 6.76 и 5.22 соответственно).

Построим совмещенную гистограмму.

```python
sns.distplot(action_user_scores['user_score'], label='Пользовательские оценки для игр жанра Action', bins=sports_user_scores['user_score'].nunique()).set_title('Распределение пользовательских оценок для игр жанров Action и Sports');
sns.distplot(sports_user_scores['user_score'], label='Пользовательские оценки для игр жанра Sports', bins=sports_user_scores['user_score'].nunique());
plt.axvline(action_user_scores['user_score'].mean(), color='b', linestyle='--', label='Средняя оценка для игр жанра Action');
plt.axvline(sports_user_scores['user_score'].mean(), color='r', linestyle='--', label='Средняя оценка для игр жанра Sports');
plt.legend();
```
