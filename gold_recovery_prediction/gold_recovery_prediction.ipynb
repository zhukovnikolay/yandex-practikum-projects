### Обработка пропусков

В задании была подсказка, что соседние по времени параметры часто похожи. Звучит как наводка на принцип заполнения пропусков. Попробуем это подтвердить. Напишем функцию, которая выводит на экран первое наблюдение с пропуском и по два наблюдения до и после него.

```python
def print_nan_values(data, feat):
    """Функция выводит на экран первое встреченное наблюдение с пропуском 
    и по два наблюдения до и после него"""
    index_with_nan = data[data[feat].isna()].index.tolist()[0]
    display(data[[feat]].loc[index_with_nan - pd.to_timedelta('2 hours') : index_with_nan + pd.to_timedelta('2 hours')])
```

```python
# определим колонки с пропусками
columns_with_nan_list = list(data_train.columns[data_train.isna().sum() > 0])
```

```python
# посмотрим на соседей данных с пропусками для каждого признака
for column in columns_with_nan_list:
    print(f'Признак {column}:')
    print(f'Границы признака: [{data_train[column].min():.1f}, {data_train[column].max():.1f}]')
    print_nan_values(data_train, column)
    print(50 * '_')
```

С учетом границ признаков (соседи имеют схожие значения и находятся далеко от границ и среднего), а также характера наблюдений во времени, заполнение ближайшим (предыдущим доступным значением - метод ffill) в большинстве случаев кажется подходящим способом обработки пропусков. Применим указанный метод.

```python
data_train = data_train.fillna(method='ffill')
data_test = data_test.fillna(method='ffill')
```

```python
# проверяем
data_train.isna().sum().sum(), data_test.isna().sum().sum()
```

### Проверка корректности расчета эффективности обогащения
Нам необходимо оценить корректность расчета эффективности обогащения.

```python
def calc_recovery(row):
    f = row['rougher.input.feed_au']
    c = row['rougher.output.concentrate_au']
    t = row['rougher.output.tail_au']
    return (c * (f - t)) / (f * (c - t)) * 100
```

```python
# посчитаем метрику
mae(data_train['rougher.output.recovery'], data_train.apply(calc_recovery, axis=1))
```

Эффективность обогащения расчитана корректно, метрика очень низкая.

## Модель

Нам необходимо предсказать два целевых значения: коэффициент восстановления золота из золотосодержащей руды на стадии черновой и финальной очисток. Для стадии черновой очистки мы используем только признаки, которые могли быть известны к этому моменту (rougher.input и rougher.state), чтобы избежать утечки данных, для стадии финальной - все признаки.

Но сначала напишем функцию для расчета метрики.

### Метрики

```python
def sMAPE(y_valid, y_pred):
    y_valid = pd.Series(y_valid)
    y_pred = pd.Series(y_pred)
    return np.mean(abs(y_valid - y_pred) / ((abs(y_valid) + abs(y_pred)) / 2)) * 100
```

```python
def final_sMAPE(sMAPE_rougher, sMAPE_final):
    return 0.25 * sMAPE_rougher + 0.75 * sMAPE_final
```

Для выбора лучшей модели для этапа флотации мы будем руководствоваться следующими вводными:
- в списке признаков не должно быть тех, которые относятся к последующим этапам, чтобы избежать data leakage;
- рассмотрим четыре набора признаков: все признаки, без размера гранул, без состояний флоатационных установок, и без обоих перечисленных наборов;
- рассмотрим три модели (линейная регрессия, случайный лес и градиентный бустинг на основе CatBoostRegressor);
- для оценки качества моделей используем кастомный скоринг на основе созданной нами функции sMAPE;
- адекватность моделей проверим при сравнении ее с DummyRegressor с методом "mean";
- при похожих результатах для базовых моделей проведем подбор гиперпараметров для выбора лучшей.
___

Подготовим упомянутые списки признаков

```python
float_columns = [column for column in data_train.columns if re.match(r'rougher.input|rougher.state', column)]
float_columns_no_size = float_columns.copy()
float_columns_no_size.remove('rougher.input.feed_size')
float_columns_no_high_corr = [column for column in float_columns if column not in high_corr_feats]
float_columns_no_size_no_high_corr = [column for column in float_columns_no_size if column not in high_corr_feats]
float_columns, float_columns_no_size, float_columns_no_high_corr, float_columns_no_size_no_high_corr
```

Признаки выбраны.

Создадим scorer для кросс-валидации модели с кастомной метрикой sMAPE.

```python
smape_scorer = make_scorer(sMAPE, greater_is_better=False)
```

Подготовим список моделей и четыре набора фичей.

```python
feats_grid = {'all_feats': float_columns,
              'no_size': float_columns_no_size,
              'no_high_corr': float_columns_no_high_corr,
              'no_size_no_high_corr': float_columns_no_size_no_high_corr}
```

```python
models_dict = {'dummy_regressor': DummyRegressor(strategy='mean'),
               'linear_regression': LinearRegression(n_jobs=-1),
               'random_forest': RandomForestRegressor(max_depth=5, n_estimators=5, random_state=22, n_jobs=-1),
               'catboost': CatBoostRegressor(random_state=22, silent=True)}
```

```python
def models_feats_grid_search(models, feats, X, y):
    """Функция перебирает модели и фичи и возвращает их усредненный sMAPE 
       по результатам кросс-валидации"""
    scores = {}
    for model_name, model in models.items():
        for feats_type, feats_list in feats.items():
            scores[model_name + '_' + feats_type] = np.abs(np.mean(cross_val_score(model, 
                                                                   X[feats_list],
                                                                   list(y),
                                                                   cv=5,
                                                                   scoring=smape_scorer)))
    return scores
```
