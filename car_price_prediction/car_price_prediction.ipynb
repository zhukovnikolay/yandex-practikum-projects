Подготовим функции для исследования признаков

```python
# функция построения гистограммы с медианой, средним и модой
def my_hist(data, column, bins=10):
    feat_median = data[column].median()
    feat_mean = data[column].mean()
    feat_mode = data[column].mode()[0]
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.hist(data[column], bins=bins)
    ax.set_title(column)
    ax.axvline(feat_median, color='g', linestyle='--', label='медиана')
    ax.axvline(feat_mean, color='r', linestyle='--', label='среднее')
    ax.axvline(feat_mode, color='y', linestyle='--', label='мода')
    ax.legend()
    plt.show()
```

```python
# функция, позволяющая исследовать признак
def feature_stat(data, column, need_hist=False, hist_bins=10, print_low=False, print_high=False, print_nan=False):
    missed_count = data[column].isna().sum()
    feat_mode = data[column].mode()[0]
    feat_stats = data[column].describe()
    if data[column].dtype not in ['object', 'bool']:
        print(f'Признак {column}\n{100 * "_"}\n'
              f'Кол-во пропусков (доля): {missed_count} ({missed_count / data.shape[0]:.1%})\n'
              f'Кол-во уникальных значений: {data[column].nunique()}\n'
              f'Минимальное значение: {feat_stats["min"]}\n'
              f'Максимальное значение: {feat_stats["max"]}\n'
              f'Медиана: {feat_stats["50%"]}\n'
              f'Среднее: {feat_stats["mean"]}\n'
              f'Мода: {feat_mode}\n'
              f'Тип данных: {data[column].dtype}\n' + 100 * '_')
    else:
        print(f'Признак {column}\n{100 * "_"}\n'
              f'Кол-во пропусков (доля): {missed_count} ({missed_count / data.shape[0]:.1%})\n'
              f'Кол-во уникальных значений: {data[column].nunique()}\n'
              f'Мода: {feat_mode}\n'
              f'Тип данных: {data[column].dtype}\n' + 100 * '_')
    if need_hist:
        print('Гистограмма')
        my_hist(data, column, bins=hist_bins)
        print(100 * '_')
    if print_low:
        print(f'Пять самых низких значений')
        display(data.sort_values(column).head(5))
        print(100 * '_')
    if print_high:
        print(f'Пять самых высоких значений')
        display(data.sort_values(column, ascending=False).head(5))
        print(100 * '_')
    if print_nan:
        print(f'Пять случайных объектов с пропусками')
        try:
            display(data[data[column].isna()].sample(5))
        except:
            print('При выполнении произошла ошибка. Возможно, в данных нет пропусков')
        print(100 * '_')
```

Наиболее разумным кажется заполнять пропуски для типа кузова аналогично автомобилям той же модели и бренда. Единственная потенциальная проблема - наличие пропусков в признаке модель. Сгруппируем данные по бренду и модели и посмотрим на типы кузовов.

```python
data.groupby(['brand', 'model', 'vehicle_type'], as_index=False)['price'].count().head(11)
```

Можно заметить, что для одной модели в объявлениях могут присутствовать несколько типов кузова. Это может быть как ошибкой при заполнении анкеты, так и особенностями моделей в зависимости от их года выпуска (в разные (да и в одни те же) годы одни и те же модели могли выпускаться в разных кузовах). Проверим наше предположение на модели Alfa Romeo 147.

Согласно данным Википедии (https://ru.wikipedia.org/wiki/Alfa_Romeo_147) данная модель выпускалась в кузове хэтчбек, таким образом, несмотря на некоторую уверенность в корректности второго предположения, более явно подтверждается версия с ошибками в типе кузова при заполнении анкет. Для упрощения работы с данными, для каждой модели оставим только один тип кузова - наиболее часто указываемый в объявлениях. Некорректные данные и пропуски заполним самым частым типом кузова.

```python
# подготовим справочник с самыми популярными типами кузова по брендам и моделям
most_freq_model_vehicle_type = data.groupby(['brand', 'model'])['vehicle_type'].value_counts()
```

```python
most_freq_model_vehicle_type = most_freq_model_vehicle_type.groupby(level=[0, 1], group_keys=False).nlargest(1)
```

```python
most_freq_model_vehicle_type = pd.DataFrame(most_freq_model_vehicle_type).\
                                                            rename(columns={'vehicle_type': 'count'}).reset_index()
most_freq_model_vehicle_type
```

```python
# подготовим справочник самого частотного типа кузова для бренда
most_freq_brand_vehicle_type = data.groupby(['brand'])['vehicle_type'].describe()['top']
most_freq_brand_vehicle_type
```

```python
def change_vehicle_type(row):
    """
    Если в модели пропуск, заполняем модой по бренду, если нет - заполняем из справочника
    """
    if pd.isna(row.model):
        row['vehicle_type'] = most_freq_brand_vehicle_type[row.brand]
    else:
        row['vehicle_type'] = most_freq_model_vehicle_type.loc[((most_freq_model_vehicle_type.brand == row.brand) & 
                                                                (most_freq_model_vehicle_type.model == row.model)), 
                                                                'vehicle_type'].values[0]
    return row
```

```python
# унифицируем типы кузова, создав признак-флаг о том, что в данные вносились изменения
data['vehicle_type_nan_flag'] = 0
data.loc[data.vehicle_type.isna(), 'vehicle_type_nan_flag'] = 1
data.loc[data.vehicle_type.isna(), 'vehicle_type'] = data.apply(lambda x: change_vehicle_type(x), axis=1)
```

Необходимо определить явные выбросы.
Попробуем найти их через 25 и 75 квантиль + 1.5 * интерквартильное расстояние.

```python
def check_anomalies(data, feat):
    q_25 = data[feat].quantile(0.25)
    q_75 = data[feat].quantile(0.75)
    iqr = q_75 - q_25
    return (q_25 - 1.5 * iqr, q_75 + 1.5 * iqr)
```

```python
print(f'Границы аномалий для registration_year: {check_anomalies(data, "registration_year")}')
```

Необходимо дополнительно немного скорректировать верхнюю границу. Год регистрации не может быть позже года парсинга данных (date_crawled). Посмотрим на максимальное значение года парсинга.

```python
def create_geo_cat(row):
    if (row['postal_code'] >= 1067 
        and row['postal_code'] <= 4889) or (row['postal_code'] >= 8056
                                            and row['postal_code'] <= 9669):
        row['geo_cat'] = 'Saxony'
    elif (row['postal_code'] >= 14461 and row['postal_code'] <= 17326):
        row['geo_cat'] = 'Brandenburg'
    elif (row['postal_code'] >= 6108 
          and row['postal_code'] <= 6928) or (row['postal_code'] >= 38820
                                            and row['postal_code'] <= 39649):
        row['geo_cat'] = 'Saxony-Anhalt'
    elif (row['postal_code'] >= 7318 
          and row['postal_code'] <= 7989) or (row['postal_code'] >= 98527
                                            and row['postal_code'] <= 99998):
        row['geo_cat'] = 'Thuringia'
    elif (row['postal_code'] >= 10115 and row['postal_code'] <= 14532):
        row['geo_cat'] = 'Berlin'
    elif (row['postal_code'] >= 17033 and row['postal_code'] <= 19417):
        row['geo_cat'] = 'Mecklenburg-Vorpommern'
    elif (row['postal_code'] >= 20038 
          and row['postal_code'] <= 21149) or (row['postal_code'] >= 22041
                                            and row['postal_code'] <= 22769):
        row['geo_cat'] = 'Hamburg'
    elif (row['postal_code'] >= 21217 
          and row['postal_code'] <= 21789) or (row['postal_code'] >= 26121
                                            and row['postal_code'] <= 38729) or (row['postal_code'] >= 49074
                                                                                 and row['postal_code'] <= 49849):
        row['geo_cat'] = 'L.Saxony'
    elif (row['postal_code'] >= 22844 and row['postal_code'] <= 25999):
        row['geo_cat'] = 'Schleswig-Holstein'
    elif (row['postal_code'] >= 27568 
          and row['postal_code'] <= 27580) or (row['postal_code'] >= 28195
                                            and row['postal_code'] <= 28779):
        row['geo_cat'] = 'Bremen'
    elif (row['postal_code'] >= 32049 
          and row['postal_code'] <= 33829) or (row['postal_code'] >= 40196
                                            and row['postal_code'] <= 54585) or (row['postal_code'] >= 58084
                                                                                 and row['postal_code'] <= 59969):
        row['geo_cat'] = 'N.Rhine-Westphalia'
    elif (row['postal_code'] >= 34117 
          and row['postal_code'] <= 37299) or (row['postal_code'] >= 60306
                                            and row['postal_code'] <= 63699) or (row['postal_code'] >= 64283
                                                                                 and row['postal_code'] <= 65936):
        row['geo_cat'] = 'Hesse'
    elif (row['postal_code'] >= 54290 
          and row['postal_code'] <= 57648) or (row['postal_code'] >= 66849
                                            and row['postal_code'] <= 67829):
        row['geo_cat'] = 'Rhineland-Palatinate'
    elif (row['postal_code'] >= 66041 and row['postal_code'] <= 66839):
        row['geo_cat'] = 'Saar'
    elif (row['postal_code'] >= 68131 
          and row['postal_code'] <= 79879) or (row['postal_code'] >= 88212
                                            and row['postal_code'] <= 89198):
        row['geo_cat'] = 'Baden-Wurttemberg'
    elif (row['postal_code'] >= 80331 and row['postal_code'] <= 97909):
        row['geo_cat'] = 'Bavaria'
    else:
        row['geo_cat'] = 'Other'
    return row
```

На первый взгляд можно рассмотреть три способа обработки такой аномалии:
1. Установка минимально допустимой цены автомобиля для всех автомобилей с ценой ниже допустимой.
2. Установка цены методом ближайших соседей для слишком низких цен.

Все три способа обработки аномалоии требуют определения некой допустимой цены. При этом, у первого подхода есть существенный минус - используя его мы можем установить очень низкую цену для достаточно дорогих автомобилей, по которым были ошибки в выгрузке или которые продавались через аукцион.

### Подготовка тренировочной и тестовой выборок

Нам необходимо подготовить два комплекта данных - для обучения линейных моделей с OHE-кодированием и для бустинга (с преобразованием типов данных object в category).

```python
data_ohe = data.copy()
data_ohe = pd.get_dummies(data_ohe, drop_first=True)

data_cat = data.copy()
data_cat = data_cat.astype({"vehicle_type":'category',
                            "gearbox":'category',
                            "model":'category',
                            "fuel_type":'category',
                            "brand":'category',
                            "not_repaired":'category',
                            "geo_cat": 'category'})
```

```python
data_cat.shape
```

6. Удалили незначащие признаки: дата, почтовый индекс, количество фоток.
7. Провели подготовку данных для обучения и проверки моделей (разделили выборку на тренировочную и тестовую, подготовили признаки в зависимости от типа модели).
